```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.layers import LSTM, Dense,Dropout
import keras_tuner as kt
df
```

```python
feature=['Close Price','High Price','Low Price','Open Price','volume']
df_model=df[feature].dropna()
scaler=StandardScaler()
df_model_scaled=scaler.fit_transform(df_model)
lookback=5
def create_sequence(data,lookback):
    x,y=[],[]
    for i in range(lookback,len(data)):
        x.append(data[i-lookback:i])
        y.append(data[i,-1])
    return np.array(x),np.array(y)
x,y=create_sequence(df_model_scaled,lookback)
num_trained=int(0.8*(len(x)))
x_train,x_test=x[:num_trained],x[num_trained:]
y_train,y_test=y[:num_trained],y[num_trained:]
def build_model(hp):
    model=Sequential()
    model.add(LSTM(hp.Int('units',min_value=50,max_value=210,step=16)
                   ,return_sequences=False,
                   input_shape=(x.shape[1],x.shape[2])))
    model.add(Dropout(hp.Float('Dropout',0.1,0.5,step=0.1)))
    model.add(Dense(1))
    model.compile(optimizer='Adam',loss='mse')
    return model
tuner=kt.RandomSearch(build_model,
                      objective='loss',
                      max_trials=5,
                      executions_per_trial=1)
tuner.search(x_train, y_train, epochs=20, batch_size=32)
best_hp=tuner.get_best_hyperparameters(1)[0]
best_model=tuner.get_best_models(num_models=1)[0]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, TimeSeriesSplit, GridSearchCV,KFold
from skopt import BayesSearchCV
from xgboost import XGBRegressor
from skopt.space import Real, Integer
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
pipe=Pipeline([('scaler',StandardScaler()),
              ('estimator',XGBRegressor)])
```

```python
cv_strategy=TimeSeriesSplit(n_splits=10)
para_space={'estimator__n_estimators':Integer(100,200),
            'estimator__max_depth':Integer(3,7),
            'estimator__learning_rate':Real(0.01,0.1)}
```

```python

pipe=Pipeline([('scaler',StandardScaler()),
              ('estimator',XGBRegressor())])
cv_strategy=TimeSeriesSplit(n_splits=10)
para_space={'estimator__n_estimators':Integer(100,200),
            'estimator__max_depth':Integer(3,7),
            'estimator__learning_rate':Real(0.01,0.1,prior='log-uniform')}
bayes=BayesSearchCV(estimator=pipe,
                    search_spaces=para_space,
                    n_iter=10,
                    cv=cv_strategy,
                    scoring='neg_mean_squared_error')
bayes.fit(x_train,y_train)
```

```python
bayes.best_score_
```

```python

pipe=Pipeline([('scaler',StandardScaler()),
              ('estimator',XGBRegressor())])
cv_strategy=TimeSeriesSplit(n_splits=10)
para_space={'estimator__n_estimators':[100,200],
            'estimator__max_depth':[3,7],
            'estimator__learning_rate':[0.1,0.2,0.3]}
grid=GridSearchCV(estimator=pipe,
                   param_grid=para_space,
                   cv=cv_strategy,
                   scoring='neg_mean_squared_error')
grid.fit(x_train,y_train)
grid.best_score_
```
